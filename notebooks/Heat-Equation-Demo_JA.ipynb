# Heat map

First here is the schema for open circular port graphs.



We will build a composition syntax that looks like this.
- $B$ - boxes
- $P$ - not shown
- $W$ - arrows
- $Q$ - outer ports, blue dots



```julia
# Pick a composition syntax

W = 10; H = 20  # we'll create a grid with width 10 and height 20

# initialize the instance
stencil = OpenCPortGraph()

# add the boxes
boxes = reshape(add_parts!(stencil, :Box, W*H), H, W) 

ports = map(i -> add_parts!(stencil, :Port, 4,  box = i), boxes)

# add up/down wires
map(view(boxes, 1:(H - 1), 1:W)) do b
    add_part!(stencil, :Wire, src = ports[b][3], tgt = ports[b + 1][1])
    add_part!(stencil, :Wire, tgt = ports[b][3], src = ports[b + 1][1])
end

# add left/right wires
map(view(boxes, 1:H, 1:(W-1))) do b
    add_part!(stencil, :Wire, src = ports[b][2], tgt = ports[b+H][4])
    add_part!(stencil, :Wire, tgt = ports[b][2], src = ports[b+H][4])
end

# add outer ports
add_parts!(stencil, :OuterPort, W, con = map(i -> ports[1,i][1], 1:W))
add_parts!(stencil, :OuterPort, W, con = map(i -> ports[H,i][3], 1:W))
add_parts!(stencil, :OuterPort, H, con = map(i -> ports[i,1][4], 1:H))
add_parts!(stencil, :OuterPort, H, con = map(i -> ports[i,W][2], 1:H))

stencil
```

Next we have to choose primitive models to live in each box. The dynamics are discrete that will take four inputs (the states of its four neighbors) and have a single internal state. That state will be "read-out" to it's four neighbors.

```julia
# Choose a primitive model. We will fill each box with the same primitive model 

point_model = DiscreteMachine{Float64}(
    4, # number of inputs
    1, # number of states
    4, # number of outputs
    (u,x,p,t) -> [p.γ * (sum(x) - 4*u[1]) + u[1]], # dynamics: u - the state vector, x - the input vector
    u -> repeat(u, 4) # the read out function
)
```

```julia
# Compose!
# Apply the point model to each box of the stencil

heat_eq = oapply(stencil, point_model)
```

```julia
# Solve and plot!

# initial coundition
u0 = zeros(Float64, W*H) 

# boundary condition
xs = 100*vcat(ones(W), zeros(W), zeros(H), zeros(H)) # order: bottom, top, left, right

# parameters
alpha = 2
delta_x = 1

delta_t = (delta_x ^ 2)/(4 * alpha)
gamma = (alpha * delta_t) / (delta_x ^ 2)

params = LVector(γ = gamma)
nsteps = 1000

# find the trajectory
traj = trajectory(heat_eq, u0, xs, params, nsteps)
```

```julia
# ... I'm sure there are better ways to do this ...

ts = [2, 11, 21, 31, 41] # time steps to plot


hms = map(ts) do t 
    heatmap(reshape(traj[t], (H, W)), clim = (0, 100.0))
end


plot(hms..., layout = (1,5), colorbar = false, 
       title = ["t = 1" "t = 10" "t = 20" "t = 30" "t = 40"], titlefont = font(8))
plot!(size=(W*50,H*10))
plot!(xticks = false)
plot!(yticks = false)
```

# Adding advection to the system

```julia
# Choose a primitive model. We will fill each box with the same primitive model 

# We will be making modifications to encode advection along with the diffusion
# Via trial and error the order of ports appears to be as follows
#
#                 1
#                 |
#            2---------4
#                 |
#                 3

point_model = DiscreteMachine{Float64}(
    4, # number of inputs
    1, # number of states
    4, # number of outputs
    (u,x,p,t) -> [p.γ * (sum(x) - 4*u[1]) + u[1] -
                  (p.uₓ/p.Δx) * (-1) * (u[1] - x[2]) - 
                  (p.uᵥ/p.Δx) * (-1) * (u[1] - x[3])], # dynamics: u - the state vector, x - the input vector
    u -> repeat(u, 4) # the read out function
)
```

```julia
# Compose!
# Apply the point model to each box of the stencil

heat_eq = oapply(stencil, point_model)
```

```julia
# Solve and plot!

# initial coundition
u0 = zeros(Float64, W*H) 

# boundary condition
xs = 100*vcat(ones(W), zeros(W), zeros(H), zeros(H)) # order: bottom, top, left, right

# parameters
alpha = 2
delta_x = 1
u_x = 0.1
u_y = 0.2

delta_t = (delta_x ^ 2)/(4 * alpha)
gamma = (alpha * delta_t) / (delta_x ^ 2)

params = LVector(γ = gamma, uₓ = u_x, uᵥ = u_y , Δx = delta_x )
nsteps = 1000

# find the trajectory
traj = trajectory(heat_eq, u0, xs, params, nsteps)
```

```julia
# ... I'm sure there are better ways to do this ...

ts = [2, 11, 21, 31, 1001] # time steps to plot


hms = map(ts) do t 
    heatmap(reshape(traj[t], (H, W)), clim = (0, 100.0))
end


plot(hms..., layout = (1,5), colorbar = false, 
       title = ["t = 1" "t = 10" "t = 20" "t = 30" "t = 1000"], titlefont = font(8))
plot!(size=(W*100,H*20))
plot!(xticks = false)
plot!(yticks = false)
```
